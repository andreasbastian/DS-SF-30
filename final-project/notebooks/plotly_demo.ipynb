```python
# graphs a series, where keys are states
def plotlyMap(series, title):
    global data, layout
    
    scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
    
    data = [ dict(
            type='choropleth',
            colorscale=scl,
            locations = series.keys(),
            z = series.values.astype(float),
            locationmode = 'USA-states',
            #text = seq['text'],
            marker = dict(
                line = dict (
                    color = 'rgb(255,255,255)',
                    width = 1
                ) ),
            colorbar = dict(
                title="",
                )
            ) ]

    layout = dict(
            title = title,
            geo = dict(
                scope='usa',
                projection=dict( type='albers usa' ),
                showlakes = False,
                lakecolor = 'rgb(255, 255, 255)'
            ),
        )
    
    fig = dict(data=data, layout=layout)
    iplot(fig)
```

```python
employeesByState = df[ ['employees', 'state'] ].groupby('state').sum() 
emp = employeesByState.drop(['British Columbia', 'Alberta', 'Chihuahua',
                       'Coahuila', 'Estado de Mexico', 'Manitoba',
                       'Nuevo Leon', 'Ontario', 'Puebla', 'Quebec'])
```

```python
plotlyMap(emp['employees'], "Foundry Workers by State")
```
