```python
# graphs a series, where keys are states
#couretesy Max Bates
def plotlyMap(series, title):
    global data, layout
    
    scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
    
    data = [ dict(
            type='choropleth',
            colorscale=scl,
            locations = series.keys(),
            z = series.values.astype(float),
            locationmode = 'USA-states',
            #text = seq['text'],
            marker = dict(
                line = dict (
                    color = 'rgb(255,255,255)',
                    width = 1
                ) ),
            colorbar = dict(
                title="",
                )
            ) ]

    layout = dict(
            title = title,
            geo = dict(
                scope='usa',
                projection=dict( type='albers usa' ),
                showlakes = False,
                lakecolor = 'rgb(255, 255, 255)'
            ),
        )
    
    fig = dict(data=data, layout=layout)
    iplot(fig)


plotlyMap(emp['employees'], "Foundry Workers by State")
```

We need to break up the String lists into something more manageable and are going to use a logisitic regression below, so use one-hot encoding to decompose casting processes, value add processes, and alloys.

```python
dfDummies = df

#let's break the categorical variables down with one-hot encoding:
dfDummies = dfDummies.join(df['casting_processes'].str.get_dummies(sep=',').rename(columns=lambda x: 'process_' + str(x)))
dfDummies = dfDummies.join(df['value_added_processes'].str.get_dummies(sep=',').rename(columns=lambda x: 'valAdd_' + str(x)))
dfDummies = dfDummies.join(df['metals'].str.get_dummies(sep=',').rename(columns=lambda x: 'alloy_' + str(x)))
dfDummies.drop(['casting_processes', 'value_added_processes', 'metals'], axis = 1, inplace = True)

dfProcesses = df['casting_processes'].str.get_dummies(sep=',')#.rename(columns=lambda x: 'process_' + str(x))
dfAlloys = df['metals'].str.get_dummies(sep=',')#.rename(columns=lambda x: 'alloy_' + str(x))
dfValAdd = df['value_added_processes'].str.get_dummies(sep=',')#.rename(columns=lambda x: 'valAdd_' + str(x))
```

Now that processes are broken up, let's count them up and see which ones are the most widely used.

```python
processCounts = {}
for col in dfProcesses.columns:
    processCounts[str(col)] = dfProcesses[str(col)].sum()
#     dfProcesses.process_Continuous.value_counts()
```

```python
y = sorted(processCounts.items(), key=operator.itemgetter(1))
labels = []
vals = []
for i in y:
    labels.append(i[0].replace('process_',''))
    vals.append(i[1])

x = range(len(vals))
width = 1/1.5
plt.bar(x, vals, width, color="blue")

plt.title('Process Popularity Among AFS Members')
plt.xlabel('Casting Process')
plt.xticks(np.arange(len(processCounts)) + width / 2, labels, rotation=90)
plt.ylabel('Foundries Offering Process')
plt.savefig("processHistogram.png")

plt.show()
```

## Cassification:  Investment Casting or Sand Casting?

Based on what alloys a foundry uses, can we predict whether it is an investment or a sand foundry?  Can we predict based on processes whether a foundry is ferrous or non-ferrous?


```python
#how many IC foundries are listed?
print "Mean investment casting foundry size:  " + str(dfDummies[dfDummies['process_Investment Casting']== 1]['employees'].mean())
print "Mean non-investment casting foundry size:  " + str(dfDummies[dfDummies['process_Investment Casting']== 0]['employees'].mean())
```

Format and label the data for the logistic regression:

```python
dfDummies.sort_values(by = 'process_Investment Casting', inplace = True)
X = dfAlloys
c = dfDummies['process_Investment Casting']
print c 
```

```python
model_IC = linear_model.LogisticRegression().fit(X,c)
print model_IC.coef_
print model_IC.intercept_
```

Score the model to see how effective it is (just on the training set):

```python
model_IC.score(X, c)
#without sorting IC==1, get 0.95522388059701491
#with sorting, get 0.92835820895522392
```

Plot the model performance:


```python
logit_hat_IC = model_IC.intercept_ + \
    (X * model_IC.coef_[0]).sum(axis = 1)
logit_hat_IC.plot()
```

```python
odds_hat_IC = np.exp(logit_hat_IC)
p_hat_IC = odds_hat_IC / (1 + odds_hat_IC)
p_hat_IC.plot()
```

See how the model performs-- any false positives or false negatives?

```python
c_hat_IC = p_hat_IC.round()
pd.crosstab(c_hat_IC,
    c,
    rownames = ['Hypothesized Class'],
    colnames = ['True Class'])
```
